```python
import pandas as pd

category_data = pd.read_csv("log_parsing.csv")
category_data
```

```python
# 정규화
category_data["stayTime"]=category_data["stayTime"]/category_data["stayTime"].sum()
category_data
```

```python
category_data = category_data.pivot_table("stayTime",index="user_Num",columns="category")
category_data.fillna(0,inplace=True)
category_data
```

```python
corr = category_data.corr(method="pearson")
corr
```

```python
category_based = pd.DataFrame(data=corr,index = category_data.columns,columns=category_data.columns)
category_based.to_csv("pearson_pivot_table.csv")
category_based
```

```python
def get_category(category):
    return category_based[category].sort_values(ascending=False)[:4] # rank 3 
    #0.5 이상 + random category -> 총 3개
    
```

```python
get_category("4050") 
```
