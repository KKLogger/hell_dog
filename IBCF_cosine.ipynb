```python
import pandas as pd

category_data = pd.read_csv("log_parsing.csv")
category_data.info()
```

```python
category_data = category_data.pivot_table("stayTime",index="category",columns="user_Num")
category_data.fillna(0,inplace=True)
category_data.to_csv("pivot_table.csv")
category_data
```

```python
from sklearn.metrics.pairwise import cosine_similarity

category_based = cosine_similarity(category_data)
category_based
```

```python
category_based = pd.DataFrame(data=category_based,index = category_data.index,columns=category_data.index)
category_based.to_csv("cosine_pivot_table.csv")
category_based
```

```python
def get_category(category):
    return category_based[category].sort_values(ascending=False)[:4] # rank 3 
    #0.5 ì´ìƒ + random category -> ì´ 3ê°œ
    
```
