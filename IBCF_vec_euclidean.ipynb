```python
import pandas as pd

category_data = pd.read_csv("log_parsing.csv")
category_data
```

```python
# 정규화
category_data["stayTime"]=category_data["stayTime"]/category_data["stayTime"].sum()
category_data
```

```python
category_data = category_data.pivot_table("stayTime",index="category",columns="user_Num")
category_data.fillna(0,inplace=True)
category_data
```

```python
from sklearn.metrics.pairwise import euclidean_distances

category_based = euclidean_distances(category_data)
category_based
```

```python
category_based = pd.DataFrame(data=category_based,index = category_data.index,columns=category_data.index)
category_based.to_csv("vec_euclidean_pivot_table.csv")
category_based
```

```python
def get_category(category):
    return category_based[category].sort_values(ascending=True)[:4] # rank 3 
    #0.5 이상 + random category -> 총 3개
    
```

```python
get_category("wedding") 
```
